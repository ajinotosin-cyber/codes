```python
def calculate(num1, num2, operation):
    if operation == "add":
        return num1 + num2
    if operation == "subtract":
        return num1 - num2
    if operation == "multiply":
        return num1 * num2
    if operation == "divide":
        if num2 == 0:
           return "Error: Division by zero"
    return num1 / num2
```

```python
print(calculate(12, 4, "add"))
print(calculate(12, 4, "subtract"))
print(calculate(12, 4, "multiply"))
print(calculate(12, 0, "divide"))
```

```python
def greet_person(*args, **kwargs):
    if not args and not kwargs:
        print("Hello, Guest! Welcome to Python!")
        return
    for name in args:
        if "title" in kwargs:
            print(f"Hello, {kwargs['title']} {name}! Welcome to Python!")
        else:
            print(f"Hello, {name}! Welcome to python!")
```

```python
greet_person()
```

```python
greet_person("Oluwatosin", "Deborah", "Modupeoluwa")
```

```python
greet_person("Oluwatosin", "Deborah", title = "Dr")
```
